```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
data_in_dir = r"../data/"

gw_station_fname = r"opendata_messstelle.csv"

gw_station_pfname = data_in_dir + gw_station_fname
```

```python
df = pd.read_csv(gw_station_pfname, delimiter=';', header='infer')
```

```python
df.e32
```

```python
df = df.dropna(subset=['e32'])
df = df.dropna(subset=['n32'])
```

```python
df.e32
```

```python

df.e32 = (df.e32.str[:-2]+"50").astype(int)
df.n32 = (df.n32.str[:-2]+"50").astype(int)
```

```python
df.e32
```

```python
df.dtypes
```

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.e32, df.n32), crs="EPSG:25832")
```

```python
gdf
```

```python
gdf.to_csv("test3.csv")
```
